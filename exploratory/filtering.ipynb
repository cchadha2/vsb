```python
from scipy.signal import butter, filtfilt, lfilter
```

```python
train_sig.head()
```

```python
train_sig.shape
```

nyquist = 40Hz sampling rate*0.5 for half the sampling rate

wn = 9Hz cutoff/(40Hz sampling rate*0.5 for half the sampling rate)

```python
cutoff=50
```

```python
measurements=800000
```

```python
time=0.02
```

```python
sampling_rate = measurements/time
```

```python
sampling_rate
```

```python
cutoff/(sampling_rate*0.5)
```

```python
nyquist = sampling_rate*0.5
```

```python
wn = cutoff/nyquist
```

```python
b, a = butter(3, wn, btype='lowpass')

filtered_sig = lfilter(b, a, train_sig['0'].values)
```

```python
filtered_sig.shape
```

```python
filtered_sig
```

```python
train_sig['3'] = filtered_sig
```

```python
train_sig.head(20)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.lineplot(train_sig[:800000].index, train_sig['3'][:800000])
```

```python
train_sig['4'] = train_sig['0'] - train_sig['3']
```

```python
train_sig.head()
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.lineplot(train_sig[:800000].index, train_sig['4'][:800000])
```

Looks like we're getting some constructive interference here - the timing and phase alignment aren't quite good enough for the destructive interference we're looking for. Let's instead try a high-pass filter with the same cut-off freq.

```python
b, a = butter(3, wn, btype='highpass')

filtered_sig = lfilter(b, a, train_sig['0'].values)
```

```python
filtered_sig.shape
```

```python
filtered_sig
```

```python
train_sig['5'] = filtered_sig
```

```python
train_sig.head(20)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.lineplot(train_sig[:800000].index, train_sig['5'][:800000])
```

Still not really getting rid of the sinusoidal component. Let's try upping the cutoff frequency to 5000Hz instead

```python
b, a = butter(3, 5000/nyquist, btype='highpass')

filtered_sig = lfilter(b, a, train_sig['0'].values)
```

```python
filtered_sig.shape
```

```python
filtered_sig
```

```python
train_sig['6'] = filtered_sig
```

```python
train_sig.head(20)
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.lineplot(train_sig[:800000].index, train_sig['6'][:800000])
```

This looks much more reasonable. But maybe 5000Hz is too high of a cut-off frequency? We might be losing some important PD information here. Let's keep reducing the cut-off until we start seeing the sinusoidal component again and choose a value just above this.

```python
b, a = butter(3, 2500/nyquist, btype='highpass')

filtered_sig = lfilter(b, a, train_sig['0'].values)
```

```python
filtered_sig.shape
```

```python
filtered_sig
```
